```python
def main():
    num1 = int(input("Enter Number 1: "))
    num2 = int(input("Enter Number 2: "))
    num3 = int(input("Enter Operation: "))

    def compute():
      def add(a, b):
        add = a + b
        print("the sum is: ", add)

      def mult(a, b):
        mult = a * b
        print("the product is: ", mult)

      def div(a, b):
        div = a / b
        print("the quotient is: ", div)

      def subt(a, b):
        subt = a - b
        print("the difference is: ", subt)

      def notvalid():
        print("Not a valid input!")

      if num3 == 1:
        add(num1, num2)
      elif num3 == 2:
        mult(num1, num2)
      elif num3 == 3:
        div(num1, num2)
      elif num3 == 4:
        subt(num1, num2)
      elif num3<4 or num3>1:
        notvalid()
    compute()
main()
```
